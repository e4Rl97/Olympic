# オリンピック 新規恒久6施設場所

```python
import numpy as np
import pandas as pd

import folium
from folium import plugins
```

```python
# mapオブジェクトを定義
trajMap = folium.Map(
    # 新しい地図を作るときの中央にくる緯度と経度を入れる
    location = [35.628820, 139.814210],
    # はじめに表示するときの縮尺を設定する
    zoom_start = 12
)

# 地図にMarkersを配置
# 東京アクアティクスセンター
folium.Marker(location = [35.651699, 139.815106],
              popup = "Tokyo Aquatics Centre").add_to(trajMap)
# 海の森水上競技場
folium.Marker(location = [35.606690, 139.813456],
              popup = "Sea-forest Waterway").add_to(trajMap)
# 有明アリーナ
folium.Marker(location = [35.643572, 139.794256],
              popup = "Ariake Arena").add_to(trajMap)
# カヌー・スラロームセンター
folium.Marker(location = [35.643667, 139.853314],
              popup = "Kasai Canoe Slalom Course").add_to(trajMap)
# 大井ホッケー競技場
folium.Marker(location = [35.592808, 139.753520],
              popup = "Oi Hockey Stadium").add_to(trajMap)
# 夢の島公園アーチェリー場
folium.Marker(location = [35.650136, 139.828389],
              popup = "Yumenoshima Park Archery Field").add_to(trajMap)

# trajMap.save('trajMap.html')
trajMap
```
